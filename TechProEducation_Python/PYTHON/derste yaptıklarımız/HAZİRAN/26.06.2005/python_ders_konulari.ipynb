## format yöntemi
`{}` yer tutucularını kullanarak string formatlama.


```python
first_name = 'Melisa'
age = 35
job = 'ML Engineer'
print('{0} is {1} years old. She is an {2}.'.format(first_name, age, job))
# veya indeksleri karıştırarak:
print('{2} is {1} years old. She is an {0}.'.format(job, age, first_name))
```

## Escape Sequences

```python
print('Disiplin, başarıya giden yolda en güçlü yol arkadaşınızdır;\nYetenek sizi başlatabilir,\nama disiplin sizi bitiş çizgisine ulaştırır.')
```

## Comparison Operators ve Falsy/Truthy
Karşılaştırma operatörleri: `==`, `!=`, `<`, `>`, `<=`, `>=`
Falsy değerler: `0`, `0.0`, `''`, `None`, `False`, `[]`, `{}`
Truthy: diğer tüm dolu/gerçek değerler


```python
print(5 > 3)        # True
print(5 == 5)       # True
print('' == False)  # False
if []:
    print('Boş liste truthy')
else:
    print('Boş liste falsy')
```

## Logic Expressions (Mantıksal İfadeler)
İşlem önceliği: `not` > `and` > `or`
- `and` tüm ifadelerin True olmasını ister (mükemmeliyetçi)
- `or` herhangi bir True bulduğunda True döner (iyimser)


```python
a = True
b = False
print(a and b)   # False
print(a or b)    # True
print(not a)     # False
# Öncelik örneği
print(True or False and False)  # True (and önce çalışır)
```

## Kısa ipuçları
- Kod düzenleme kısayolu: Ctrl + K + F (kod formatlama)
- Değişken isimlendirmede açıklayıcı ol
- Her örneği çalıştırarak sonucu kontrol et


```python
# Swap örneği (tuple unpacking)
a = 10
b = 20
a, b = b, a
print(f'a = {a}, b = {b}')

# İndirimli fiyat hesaplama
urun_fiyat = float(input("Lütfen ürünün fiyatını giriniz: "))
oran = float(input("İndirim oranını giriniz (%): "))
indirimli_fiyat = urun_fiyat - (urun_fiyat * oran / 100)
print(f"Ürünün %{oran} indirim sonrası fiyatı: {indirimli_fiyat} TL")
```
