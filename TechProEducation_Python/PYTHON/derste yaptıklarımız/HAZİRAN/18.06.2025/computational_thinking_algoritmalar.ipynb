### Trafik Işıkları Senaryosu
- **Decomposition:** Yeşil, sarı ve kırmızı ışık durumlarının belirlenmesi
- **Pattern Recognition:** Hangi araçların hangi ışıkta geçtiğinin tespiti
- **Abstraction:** Kırmızı ışıkta geçme durumlarının incelenmesi
- **Algorithm:** Kırmızı ışıkta geçen araçların tespiti ve ceza kaydı oluşturulması

### 1'den 100'e Toplam Problemi
- **Decomposition:** 1+2+3…+100
- **Pattern Recognition:** 1+100=101, 2+99=101, 3+98=101
- **Abstraction:** 50 adet 101
- **Algorithm:** 50 x 101

```python
toplam = sum(range(1, 101))
print('Toplam:', toplam)

n = 100
gauss = (n * (n + 1)) // 2
print('Gauss yöntemiyle toplam:', gauss)
```
