## Değişken İsimlendirme Kuralları (Kısa)

- Değişken adı harf veya `_` ile başlamalı; rakamla başlayamaz.
- Boşluk, `-`, `.`, `$` gibi özel karakterler kullanılamaz.
- Küçük/büyük harf ayrımı vardır.
- Anlamlı ve tutarlı isimlendirme tercih edilir (örn. `first_name`).


```python
# Düzeltilmiş ve geçerli değişken isimleri
name_1 = 'Melisa'
name_2 = 'Anna'
name_3 = 'Yasin'
first_name = 'George'
name_4 = 'Ceylan'
name_5 = 'Ali'
firstName = 'Cenk'   # camelCase örneği (geçerli)
name_6 = 'Selim'
first_name_2 = 'Belen'

# Yazdırma örneği
print(name_1, name_2, name_3)
print(first_name, name_4, name_5)
print(firstName, name_6, first_name_2)
```

## Yaş ile Basit Hesaplama
Örnek: `age = "40"` verildiğinde 2025 yılına göre doğum yılını hesaplama.

```python
age = "40"
birth_year_2025 = 2025 - int(age)
print(f"2025 yılına göre doğum yılı: {birth_year_2025}")

# Güncel yıla göre hesaplama (daha doğru)
from datetime import datetime
current_year = datetime.now().year
birth_year_now = current_year - int(age)
print(f"Güncel yıla göre doğum yılı: {birth_year_now}")
```

## Dairenin Alanı
Formül: alan = π * r^2
Örnek: r = 7. `pi` için `math.pi` kullanmak daha doğru sonuç verir.

```python
import math

r = 7
pi = 3.14            # ondalık için nokta kullan
alan_approx = pi * r**2
print("Yaklaşık alan (pi=3.14):", alan_approx)

# Gerçekçi hesaplama
alan_true = math.pi * r**2
print("Gerçek alan (math.pi):", alan_true)
```

## Pisagor ile Hipotenüs
Formül: c = sqrt(a^2 + b^2)  Örnek: a = 5, b = 12

```python
import math

a = 5
b = 12
c = math.sqrt(a**2 + b**2)
print(f"Hipotenüs c = {c}")
# Alternatif (üs ile)
c_alt = (a**2 + b**2)**0.5
print(f"Hipotenüs (alternatif) c = {c_alt}")
```
