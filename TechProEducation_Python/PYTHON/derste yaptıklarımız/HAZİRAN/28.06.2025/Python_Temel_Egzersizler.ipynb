# 1.  Geometrik Hesaplamalar

**Üçgen Alanı Hesaplama**

Üçgenin alanı $Alan = \frac{taban \times yukseklik}{2}$ formülü ile hesaplanır.

```python
# Veri tanımlama
base = 5
height = 10

# Algoritma: (Taban * Yükseklik) / 2
area_triangle = (base * height) / 2

print(f"Tabanı {base}, yüksekliği {height} olan üçgenin alanı: {area_triangle} cm²'dir.")
```

* Daire Çevresi Hesaplama

Dairenin çevresi $2 \times \pi \times r$ formülü ile hesaplanır.

```python
radius = 7
pi = 3 # Sabit değer olarak alınmıştır

# Çevre hesaplama
circumference = 2 * pi * radius

print(f"Yarıçapı {radius} olan dairenin çevresi: {circumference} cm'dir.")
```

# 2. Değişken Manipülasyonu (Variable Swapping)

Verilen **a=10** ve **b=20** değerlerini, yeni bir değer atamadan (veya Pythonik yöntemlerle) yer değiştirmek için en optimize yaklaşım "tuple unpacking" yöntemidir.

```python
a = 10
b = 20

# Pythonic Way: Temporary variable kullanmadan swap işlemi
a, b = b, a

print(f"Yeni Değerler -> a: {a}, b: {b}")
```

# 3. Finansal ve Birim Dönüştürücü Algoritmalar


**Alışveriş ve İndirim Hesaplama**

Burada input fonksiyonuyla kullanıcıdan veri alırken **float** ve **int** cast işlemleri zorunludur.

```python
price = float(input("Ürün fiyatını giriniz: "))
discount_rate = float(input("İndirim oranını giriniz (%): "))

# İndirimli fiyat algoritması
discounted_price = price - (price * discount_rate / 100)

print(f"Orijinal: {price} TL | İndirim: %{discount_rate} | Net: {discounted_price} TL")
```

- **Zaman Dönüşürücü (Saniyeden Saat/Dakika/Saniyeye**

Tam sayı bölme (**//**) ve mod alma (**%**) operatörlerini kullanarak hiyerarşik bir ayrıştırma yapılır.

```python
total_seconds = int(input("Saniye değerini giriniz: "))

hours = total_seconds // 3600
minutes = (total_seconds % 3600) // 60
seconds = total_seconds % 60

print(f"{total_seconds} saniye = {hours} saat, {minutes} dakika, {seconds} saniye.")
```

# 4. String Slicing (Veri Ayrıştırma)

12 haneli öğrenci numarasını belirli indeks aralıklarına göre parçalamak, veri güvenliği ve doğru raporlama için kritik bir tekniktir.

| Veri Tipi   | İndeks Aralığı | Açıklama         |
|-------------|----------------|------------------|
| Giriş Yılı  | [0:4]          | İlk 4 hane       |
| Fakülte No  | [4:6]          | 5. ve 6. hane    |
| Bölüm No    | [6:8]          | 7. ve 8. hane    |
| Öğrenim No  | [8:10]         | 9. ve 10. hane   |
| Giriş Sırası| [10:]          | Son 2 hane       |

```python
student_no = input("12 haneli öğrenci numarasını giriniz: ")

# String slicing (dilimleme)
entry_year = student_no[0:4]
faculty_code = student_no[4:6]
department_code = student_no[6:8]
study_no = student_no[8:10]
rank = student_no[10:12]

print(f"Giriş Yılı: {entry_year}\nFakülte: {faculty_code}\nBölüm: {department_code}\nSıra: {rank}")
```

## 8. Saniye Dönüştürücü (saat, dakika, saniye)
Girilen saniyeyi saat, dakika ve saniye cinsinden yazdırınız.


```python
sn = int(input('Süreyi saniye cinsinden giriniz: '))
saat = sn // 3600
kalan1 = sn % 3600
dak = kalan1 // 60
kalan_sn = kalan1 % 60
print(f'{sn} saniye = {saat} saat, {dak} dakika, {kalan_sn} saniye')
```

```python
Saniye_Girdisi = float(input('Saniyeyi Giriniz:'))
Saat = Saniye_Girdisi // 3600
Dakika = (Saniye_Girdisi - Saat *3600) // 60
Saniye = Saniye_Girdisi - (Saat * 3600 + Dakika * 60)
print(f'{Saniye_Girdisi} saniye, {Saat} saat, {Dakika} dakika, {Saniye} saniye')
```

## 9. Sıcaklık Dönüştürücü (C ↔ F) 
Kullanıcıdan Celsius veya Fahrenheit cinsinden bir sıcaklık alarak diğer birime çeviriniz.


```python
secim = input("Celsius için 'C', Fahrenheit için 'F' giriniz: ")

if secim.upper() == "C":
    c = float(input("Sıcaklığı Celsius cinsinden giriniz: "))
    f = (c * 9/5) + 32
    print(f"{c} °C = {f} °F")
elif secim.upper() == "F":
    f = float(input("Sıcaklığı Fahrenheit cinsinden giriniz: "))
    c = (f - 32) * 5/9
    print(f"{f} °F = {c} °C")
else:
    print("Geçersiz seçim!")
```
